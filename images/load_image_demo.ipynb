```python
import numpy as np
from matplotlib import pyplot as plt

import torch
import torchvision
```

```python
image_i = torchvision.io.read_image("images/BloodMNIST/image0_(label7).png")
print(f"{image_i.size() = }\n{image_i.dtype = }")
```

```python
plt.imshow(np.transpose(image_i, (1, 2, 0)))

plt.show()
```

```python
image_f = torchvision.transforms.functional.convert_image_dtype(image_i, torch.float32)
print(f"{image_f.size() = }\n{image_f.dtype = }")
```

```python
plt.imshow(np.transpose(image_f, (1, 2, 0)))

plt.show()
```

```python
image_f_reshaped = np.transpose(image_f.reshape((3, 28 * 28)).numpy(), (1, 0))

plt.hist(
    image_f_reshaped, 32, density=True, histtype="step", color=("tab:red", "tab:green", "tab:blue")
)

plt.show()
```
